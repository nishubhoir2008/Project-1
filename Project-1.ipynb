```python
#1. Create a student record system that allows adding student data (name, roll, marks) and saving the data to a file.
with open("students.txt", "a") as f:
    name = input("Enter Name: ")
    roll = input("Enter Roll No: ")
    marks = input("Enter Marks: ")
    f.write(f"{name},{roll},{marks}\n")

print("Record Saved")
```

```python
#2. Implement search functionality to find a student by roll number.
roll = input("Enter Roll No to Search: ")

with open("students.txt", "r") as f:
    found = False
    for line in f:
        data = line.strip().split(",")
        if data[1] == roll:
            print("Name:", data[0], "Marks:", data[2])
            found = True

if not found:
    print("Record Not Found")
```

```python
#3. Add functionality to update student marks.
roll = input("Enter Roll No: ")
new_marks = input("Enter New Marks: ")

lines = []

with open("students.txt", "r") as f:
    for line in f:
        data = line.strip().split(",")
        if data[1] == roll:
            data[2] = new_marks
        lines.append(",".join(data) + "\n")

with open("students.txt", "w") as f:
    f.writelines(lines)

print("Record Updated")
```

```python
#4. Delete a student record by name or roll number.
roll = input("Enter Roll No to Delete: ")

lines = []

with open("students.txt", "r") as f:
    for line in f:
        data = line.strip().split(",")
        if data[1] != roll:
            lines.append(line)

with open("students.txt", "w") as f:
    f.writelines(lines)

print("Record Deleted")
```

```python
#5. Show all student records sorted by marks.
students = []

with open("students.txt", "r") as f:
    for line in f:
        data = line.strip().split(",")
        students.append(data)

students.sort(key=lambda x: int(x[2]), reverse=True)

for s in students:
    print(s)
```

```python
#6. Display student records with marks greater than 75.
with open("students.txt", "r") as f:
    for line in f:
        data = line.strip().split(",")
        if int(data[2]) > 75:
            print(data)
```

```python
#7. Export top 3 performing students to a new file.
students = []

with open("students.txt", "r") as f:
    for line in f:
        students.append(line.strip().split(","))

students.sort(key=lambda x: int(x[2]), reverse=True)

with open("top3.txt", "w") as f:
    for s in students[:3]:
        f.write(",".join(s) + "\n")

print("Top 3 Exported")
```

```python
#8. Implement data backup feature (copying all data to backup.txt).
with open("students.txt", "r") as source:
    data = source.read()

with open("backup.txt", "w") as backup:
    backup.write(data)

print("Backup Created")
```

```python
#9. Count how many students scored more than 90 and less than 35.
above90 = 0
below35 = 0

with open("students.txt", "r") as f:
    for line in f:
        marks = int(line.strip().split(",")[2])
        if marks > 90:
            above90 += 1
        elif marks < 35:
            below35 += 1

print("Above 90:", above90)
print("Below 35:", below35)
```

```python
#10. Create a menu-driven system (Add, View, Search, Update, Delete, Exit).
while True:
    print("\n1.Add 2.View 3.Exit")
    ch = input("Enter Choice: ")

    if ch == "1":
        with open("students.txt", "a") as f:
            name = input("Name: ")
            roll = input("Roll: ")
            marks = input("Marks: ")
            f.write(f"{name},{roll},{marks}\n")
        print("Record Added")

    elif ch == "2":
        with open("students.txt", "r") as f:
            print(f.read())

    elif ch == "3":
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```
